```python
# make an environment
env_name = "MountainCar-v0"
env = make_env(env_name)
```

```python
# increase episode length from 200 to 500
env._max_episode_steps = 500

env.reset()
print("Mountain Car Environment")
plt.imshow(env.render("rgb_array"))
env.close()
```

```python
# create a n-SARSA Learning Agent
step_size   = 0.8
episode_cnt = 1000
epsilon     = 0.0
gamma       = 1.0
lambd       = 0.5

# You can use of the two tracer to see how they behave
# They wil not have much difference in limit but behave differently
# to the parameters `step_size`, `lambd`, `epsilon`.

estimator = QEstimator(env, step_size, epsilon=epsilon, lambd=lambd,
                       gamma=gamma, trace_fn=accumulating_trace)
rewards = sarsa_lambda(env, estimator, episode_cnt=episode_cnt,
                       gamma=gamma)

# plot rewards
plot_rewards("Mountain Car World", rewards,
             "Semi Grad SARSA(Î»). Replacing trace")
```

```python
# Animate learned policy
save_dir='./videos/MountainCar-v0'
generate_animation(env, estimator, save_dir=save_dir)
[filepath] = glob.glob(os.path.join(save_dir, '*.mp4'))
display_animation(filepath)
env.close()
```
